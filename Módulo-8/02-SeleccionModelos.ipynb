```python
#Bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
```

```python
#Descargamos la tabla desde el archivo CSV y la imprimimos.
tabla = pd.read_csv("https://gitlab.com/dgtic5/res/-/raw/main/aprendizajeSupervizado/Diabetes__Edited_.csv")
tabla
```

```python
#Separación de la columna I.
y = tabla["I"]
y
```

```python
#Selección del resto de datos.
x = tabla.drop(["I"],axis=1)
x
```

<h3>Muestreo aleatorio</h3>

```python
#División de los datos en conjuntos de entrenamiento y de prueba.
#(arreglo(s),tamaño de prueba, semilla)
xEntrenamiento, xPrueba, yEntrenamiento, yPrueba = train_test_split(x,y,test_size=0.2,random_state=115)
xEntrenamiento
```

```python
#Segunda partición de los datos para crear datos de validación.
xEntrenamiento, xValidacion, yEntrenamiento, yValidacion = train_test_split(xEntrenamiento,yEntrenamiento,test_size=0.15,random_state=115)
xEntrenamiento
```

```python
xValidacion
```

<h3>Muestreo estratificado</h3>

```python
#Partición de datos de manera estratificada.
xEntrenamiento, xPrueba, yEntrenamiento, yPrueba = train_test_split(x,y,test_size=0.2,random_state=115, stratify=y)
xEntrenamiento
```

```python
xEntrenamiento, xValidacion, yEntrenamiento, yValidacion = train_test_split(xEntrenamiento,yEntrenamiento,test_size=0.15,random_state=115, stratify=yEntrenamiento)
xEntrenamiento
```

<h3>Validación cruzada</h3>

```python
xEntrenamiento, xPrueba, yEntrenamiento, yPrueba = train_test_split(x,y,test_size=0.2,random_state=115, stratify=y)
kf = KFold(n_splits=5)
for i, (train_index, test_index) in enumerate(kf.split(xEntrenamiento)):
    print(f"###########################")
    print(f"Pliegue {i}:")
    print(f"  Entrenamiento: índice={train_index}")
    print(f"  Prueba: índice={test_index}")
```
