```python
#Resolver el 8-puzzle usando el algoritmo A* (50 pts)

import copy
from collections import deque

class Nodo:
    def __init__(self, tablero, g, h):
        self.tablero = tablero
        self.g = g
        self.h = h
        self.i = self.g+self.h

puzzle = [['1','2','3'],['8','0','4'],['7','6','5']]
objetivo = [['2','8','1'],['0','4','3'],['7','6','5']]

def EncontrarVacio(tablero):
    #Encontramos la posici√≥n vac√≠a.
    for i in range(3):
        for j in range(3):
            if tablero[i][j] == '0':
                yCero = i
                xCero = j
    return [yCero, xCero]

def MovimientoValido(tablero, movimiento):
    yCero, xCero = EncontrarVacio(tablero)
    #Arriba
    if movimiento == 0:
        if yCero > 0:
            return True
    #Abajo
    if movimiento == 1:
        if yCero < 2:
            return True
    #Izquierda
    if movimiento == 2:
        if xCero > 0:
            return True
    #Derecha
    if movimiento == 3:
        if xCero < 2:
            return True
    return False

def CrearHijo(tablero, movimiento):
    yCero, xCero = EncontrarVacio(tablero)
    #Movimiento hacia arriba.
    if movimiento == 0:
        tablero[yCero][xCero] = tablero[yCero-1][xCero]
        tablero[yCero-1][xCero] = '0'
    #Movimiento hacia abajo.
    if movimiento == 1:
        tablero[yCero][xCero] = tablero[yCero+1][xCero]
        tablero[yCero+1][xCero] = '0'
    #Movimiento hacia la izquierda.
    if movimiento == 2:
        tablero[yCero][xCero] = tablero[yCero][xCero-1]
        tablero[yCero][xCero-1] = '0'
    #Movimiento hacia la derecha.
    if movimiento == 3:
        tablero[yCero][xCero] = tablero[yCero][xCero+1]
        tablero[yCero][xCero+1] = '0'
    #Regresamos el tablero nuevo.
    return tablero

def Similitud(tablero, objetivo):
    similitud = 0
    for i in range(3):
        for j in range(3):
            if(tablero[i][j] == objetivo[i][j]):
                similitud += 1
    return similitud

def DistanciaManhattan(tablero,objetivo):
    distancia = 0
    for filaTablero in range(3):
        for colTablero in range(3):
            for filaObjetivo in range(3):
                for colObjetivo in range(3):
                    if(tablero[filaTablero][colTablero] == objetivo[filaObjetivo][colObjetivo]):
                        distancia += abs(filaTablero - filaObjetivo) + abs(colTablero - colObjetivo)
    return distancia

#Funci√≥n para realizar la b√∫squeda A*
def BusquedaAEstrella(inicio, objetivo):
    visitados = []
    colaPrioridad = deque([inicio])
    profundidad = 0
    while colaPrioridad:
        nodoActual = colaPrioridad.popleft()
        if nodoActual.tablero == objetivo:
            print('Se encontr√≥ el objetivo:')
            print(nodoActual.tablero)
            return True

        if nodoActual.tablero not in visitados:
            visitados.append(nodoActual.tablero)
            print(f'Visitados{len(visitados)}')
            #print('Visitados:')
            #print(visitados)
            print('\nEstamos en el nodo:')
            print(nodoActual.tablero)
            profundidad += 1
            print(f'Profundidad: {profundidad}')
            for i in range(4):
                if MovimientoValido(nodoActual.tablero, i):
                    hijoTablero = CrearHijo(copy.deepcopy(nodoActual.tablero), i)
                    hijo = Nodo(hijoTablero,profundidad,DistanciaManhattan(hijoTablero,objetivo))
                    colaPrioridad.append(hijo)
                    print('Agregando hijo:')
                    print(hijo.tablero)
        colaPrioridad = deque(sorted(list(colaPrioridad),key=lambda nodo: nodo.i))
        print()
    print(f'No se encontr√≥ el objetivo {objetivo}.')
    return False

nodoInicio = Nodo(puzzle,0,DistanciaManhattan(puzzle,objetivo))
#print(objetivo)
#print(nodoInicio.tablero)
#print(nodoInicio.g)
#print(nodoInicio.h)
BusquedaAEstrella(nodoInicio, objetivo)
```
