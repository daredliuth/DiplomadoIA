```python
import matplotlib.pyplot as plt
import math
from scipy.special import comb

x0 = 0
xn = 10
paso = 1
valoresK = []
valoresP = []
for k in range(4):
    valoresK.append(k)
    p = comb(3,k) * math.pow(0.7,k) * math.pow(0.3,3-k)
    valoresP.append(p)
plt.plot(valoresK,valoresP,'-o')
plt.xlabel('k')
plt.ylabel('P(ξ=k)')
plt.show()
print('Valores de probabilidad:')
print(valoresP)
```

```python
import matplotlib.pyplot as plt

# Datos de la distribución conjunta
xi_eta = [(0, 0), (1, 2), (1, 4), (2, 6), (2, 4), (3, 8)]
probabilidades = [0.027, 0.126, 0.063, 0.294, 0.147, 0.343]

# Crear el gráfico
xi, eta = zip(*xi_eta)
plt.scatter(xi, eta, s=[p * 1000 for p in probabilidades], alpha=0.6, color='purple')
for i, p in enumerate(probabilidades):
    plt.text(xi[i], eta[i], f"{p:.3f}", fontsize=10, ha='center')

plt.xlabel('ξ (Número de animales)')
plt.ylabel('η (Número de patas)')
#plt.title('Distribución conjunta de (ξ, η)')
plt.grid()
plt.show()
```
