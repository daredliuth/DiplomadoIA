```python
#Crear un juego de tic-tac-toe en el que dos jugadores (uno controlado por la máquina)
#jueguen de manera que la máquina utilice el algoritmo Minimax para elegir su movimiento.
#El objetivo de la máquina es maximizarsu puntuación (hacer tres en línea), mientras que el oponente
#intenta hacer lo mismo.
#Utilizar una cola como estructura de datos  para su implementación (50 pts)

import random
import copy
import math

jugador = 'x'
oponente = 'o'

def EspaciosLibres(tablero):
    espacios = []
    for i in range(3):
        for j in range(3):
            if(tablero[i][j] != jugador and tablero[i][j] != oponente ):
                espacios.append([i,j])
    return espacios

def Ganador(tablero, turno):
    for i in range(3):
        if(tablero[i][0] == tablero[i][1] == tablero[i][2] == turno ):
            return True
    for i in range(3):
        if(tablero[0][i] == tablero[1][i] == tablero[2][i] == turno ):
            return True
    if(tablero[0][0] == tablero[1][1] == tablero[2][2] == turno ):
        return True
    if(tablero[2][0] == tablero[1][1] == tablero[0][2] == turno ):
        return True
    return False

#def CambiarTurno(turno):
#    if(turno==jugador):
#        return oponente
#    else:
#        return jugador

def MostrarTablero(tablero):
    for i in range(3):
        print(tablero[i])
    print()

def Evaluar(tablero):
    if(Ganador(tablero,jugador)):
        return 10
    elif(Ganador(tablero,oponente)):
        return -10
    return 0

def Minimax(tablero,profundidad,maximizar):
    puntuacion = Evaluar(tablero)
    if(puntuacion == 10):
        return puntuacion
    if(puntuacion == -10):
        return puntuacion
    if(len(EspaciosLibres(tablero)) == 0):
        return 0
    if(maximizar):
        mejor = 1000
        for movimiento in EspaciosLibres(tablero):
            x,y = movimiento
            tablero[x][y] = jugador
            mejor = max(mejor,Minimax(tablero,profundidad+1,not maximizar))
            tablero[x][y] = ''
        return mejor
    else:
        mejor = -1000
        for movimiento in EspaciosLibres(tablero):
            x,y = movimiento
            tablero[x][y] = oponente
            mejor = min(mejor,Minimax(tablero,profundidad+1,not maximizar))
            tablero[x][y] = ''
        return mejor

def EncontrarMejorMovimiento(tablero):
    mejorValor = -1000
    mejorMovimiento = [-1,-1]
    for movimiento in EspaciosLibres(tablero):
        x,y = movimiento
        tablero[x][y] = jugador
        valorMovimiento = Minimax(tablero, 0, False)
        tablero[x][y] = ''
        if(valorMovimiento > mejorValor):
            mejorMovimiento = movimiento
            mejorValor = valorMovimiento
    #print(f'El mejor valor es: {mejorValor}')
    #print()
    return mejorMovimiento

ticTacToe = [['','',''],['','','',],['','','']]
turnoActual = jugador
ganador = False
seleccion = False
for i in range(9):
    #Humano
    if(turnoActual == jugador):
        espacios = EspaciosLibres(ticTacToe)
        seleccionJugador = input(f'Turno jugador:\nSeleccione una casilla de la lista:\n{espacios}')
        x,y = espacios[int(seleccionJugador)]
        ticTacToe[x][y] = jugador
        ganador = Ganador(ticTacToe,jugador)
        if(ganador):
            print(f'El ganador es {jugador}')
            break
        turnoActual = oponente
    else:
        #print('Turno computadora.')
        espacios = EspaciosLibres(ticTacToe)
        #x,y = random.sample(espacios,1)[0]
        x,y = EncontrarMejorMovimiento(ticTacToe)
        turnoActual = oponente
        ticTacToe[x][y] = oponente
        ganador = Ganador(ticTacToe,oponente)
        if(ganador):
            print(f'El ganador es {oponente}')
            break
        turnoActual = jugador
    MostrarTablero(ticTacToe)
MostrarTablero(ticTacToe)
if(not ganador):
    print('Empate.')
```
