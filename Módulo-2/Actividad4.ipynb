<h3>1.- f(x)=(x^4−3)^2+4 </h3>

```python
#Implementar un algoritmo de búsqueda local utilizando el recocido simulado
#para encontrar el mínimo de las siguientes funciones matemáticas.

import random
import math
import matplotlib.pyplot as plt
import numpy as np

#1.- f(x)=(x^4−3)^2+4 (20 pts)
# Definimos la función objetivo ( f(x)=(x^4−3)^2+4 )
def FuncionObjetivo(x):
    return (pow(pow(x,4) - 3,2) + 4)

def Graficar(funcion, x0, xn, paso):
    valoresX = []
    valoresY = []
    for x in np.arange(x0,xn,paso):
        valoresX.append(x)
        valoresY.append(FuncionObjetivo(x))
    plt.plot(valoresX,valoresY)
    plt.show()
    print(f'El valor mínimo local es: f(x)={np.min(valoresY)}')


def RecocidoSimulado(funcion, limiteIteraciones=1000, paso=0.01, alfa = 0.7):
    # Empezamos con un valor inicial aleatorio
    xActual = random.uniform(-10, 10)
    t=1000

    for iteracion in range(limiteIteraciones):
        # Calculamos el valor de la función en el punto actual
        valorActual = funcion(xActual)
        # Evaluamos el valor de la función en los puntos cercanos (x - paso y x + paso)
        xVecinoIzquierda = xActual - paso
        xVecinoDerecha = xActual + paso
        valorVecinoIzquierda = funcion(xVecinoIzquierda)
        valorVecinoDerecha = funcion(xVecinoDerecha)
        if valorVecinoIzquierda < valorActual:
            xVecino = xVecinoIzquierda
            valorVecino = valorVecinoIzquierda
        else:
            xVecino = xVecinoDerecha
            valorVecino = valorVecinoDerecha

        if(valorVecino < valorActual):
            xAnterior = xActual
            xActual = xVecino
        else:
            if(t>0.001):
                probabilidad = math.exp((valorVecino-valorActual)/t)
                #print(f"Probabilidad: {probabilidad}")
                if(probabilidad > 5):
                    xAnterior = xActual
                    xActual = xVecino
            else:
                break
        
        t=alfa*t;

        if(xAnterior == xActual):
            break
        
        #print(f"Iteración {iteracion + 1}: x = {xActual}, f(x) = {valorActual}")
    print(f"El mínimo se encontró en la iteración: {iteracion + 1}: x = {xActual}, f(x) = {valorActual}")    

RecocidoSimulado(FuncionObjetivo)
Graficar(FuncionObjetivo, -1.5, 1.5, 0.1)
```
