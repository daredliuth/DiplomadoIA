```python
import random
import math
import copy

#Función para generar una población inicial de cromosomas binarios de longitud fija.
def GenerarPoblacion(longitud, tamPoblacion):
    poblacionInicial = [[] for _ in range(tamPoblacion)]
    for i in range(tamPoblacion):
        for j in range(longitud):
            poblacionInicial[i].append(random.randint(0,1))
    return poblacionInicial

#Función para mapear valores binarios a decimales.
def Mapeo(cromosoma):
    valorReal = 0
    for i in range(len(cromosoma)):
        valorReal += int(int(cromosoma[-(i+1)]) * math.pow(2,i))
    return valorReal

def MapeoCaso1(cromosoma):
    valorReal = 0
    l = 0
    u = 5
    n = 3
    for i in range(len(cromosoma)):
        valorReal += int(int(cromosoma[-(i+1)]) * math.pow(2,i))
    fenotipo = l + (valorReal/(math.pow(2,n)-1)) * (u-l)
    return fenotipo

#Función para separar un cromosoma en 3 valores binarios.
def Separacion(cromosoma):
    tam = int(len(cromosoma)/3)
    valores = []

    for i in range(tam):
        valores.append('')
        valores[i] = str(cromosoma[0+(i*3)]) + str(cromosoma[1+(i*3)]) + str(cromosoma[2+(i*3)]) 
    return valores

#Función para calcular la aptitud de un cromosoma.
def Aptitud(cromosoma):
    l, w, h = Separacion(cromosoma)
    l = Mapeo(l)
    w = Mapeo(w)
    h = Mapeo(h)
    aptitud = l*w*h
    return aptitud

def AptitudCaso1(cromosoma):
    l, w, h = Separacion(cromosoma)
    l = MapeoCaso1(l)
    w = MapeoCaso1(w)
    h = MapeoCaso1(h)
    aptitud = l*w*h
    return aptitud

poblacionInicial = GenerarPoblacion(9,4)

print(f'Cromosoma: {poblacionInicial[0]}')
valorReal = Mapeo(poblacionInicial[0])
print(f'Valor decimal del cromosoma: {valorReal}')
l,w,h = Separacion(poblacionInicial[0])
print(f'l: 0b{l} = {Mapeo(l)}\nw: 0b{w} = {Mapeo(w)}\nh: 0b{h} = {Mapeo(h)}')
aptitud = Aptitud(poblacionInicial[0])
print(f'Aptitud: {aptitud}')
```
