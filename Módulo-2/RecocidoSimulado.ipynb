```python
#Implementar un algoritmo de búsqueda local utilizando el Hill Climbing con recocido simulado para encontrar el mpinimo de una función.
#f(x)=(x^4−3)^2+4

import random
import math
import copy

# Definimos la función objetivo (f(x) = (x - 3)^2 + 4)
def funcion_objetivo(x):
    return (x - 3)**2 + 4

def descenso_colinas(funcion, limite_iteraciones=1000, paso=0.1, alfa = 0.7):
    # Empezamos con un valor inicial aleatorio
    x_actual = random.uniform(-10, 10)
    t=1000

    for iteracion in range(limite_iteraciones):
        # Calculamos el valor de la función en el punto actual
        valor_actual = funcion(x_actual)
        # Evaluamos el valor de la función en los puntos cercanos (x - paso y x + paso)
        x_vecino_izquierda = x_actual - paso
        x_vecino_derecha = x_actual + paso
        valor_vecino_izquierda = funcion(x_vecino_izquierda)
        valor_vecino_derecha = funcion(x_vecino_derecha)
        if valor_vecino_izquierda < valor_actual:
            x_vecino = x_vecino_izquierda
            valor_vecino = valor_vecino_izquierda
        else:
            x_vecino = x_vecino_derecha
            valor_vecino = valor_vecino_derecha

        if(valor_vecino < valor_actual):
            x_anterior = x_actual
            x_actual = x_vecino
        else:
            if(t>0.001):
                probabilidad = math.exp((valor_vecino-valor_actual)/t)
                print(f"Probabilidad: {probabilidad}")
                if(probabilidad > 5):
                    x_anterior = x_actual
                    x_actual = x_vecino
            else:
                break
        
        t=alfa*t;

        if(x_anterior == x_actual):
            break
        
        print(f"Iteración {iteracion + 1}: x = {x_actual}, f(x) = {valor_actual}")    
    return x_actual

# Ejecutamos el descenso de colinas para encontrar el mínimo
minimo = descenso_colinas(funcion_objetivo)
print(f"El mínimo local encontrado es en x = {minimo}")
```
