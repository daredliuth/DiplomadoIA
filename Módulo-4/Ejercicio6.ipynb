```python
import matplotlib.pyplot as plt
import math
from scipy.special import comb

x0 = 0
xn = 10
paso = 1
valoresK = []
valoresP = []
for k in range(4):
    valoresK.append(k)
    p = comb(3,k) * math.pow(0.7,k) * math.pow(0.3,3-k)
    valoresP.append(p)
plt.plot(valoresK,valoresP,'-o')
plt.xlabel('k')
plt.ylabel('P(Î¾=k)')
plt.show()
print('Valores de probabilidad:')
print(valoresP)
```
