```python
#Bibliotecas
import cv2
import os
```

```python
#Funciones

#Guardas las imágenes de una carpeta en un arreglo.
def GuardarImagenesArreglo(arreglo, ruta):
    for raiz, carpetas, archivos in os.walk(ruta):
        print(f"Ruta: {ruta}")
        for archivo in archivos:
            arreglo.append(cv2.imread(ruta+archivo))

#Reescalar imágenes a un tamaño específico.
def ReescalarImagenes(imagenes):
    for i in range(len(imagenes)):
        imagenes[i] = cv2.resize(imagenes[i],(512, 512))

#Convierte las imágenes a blanco y negro.
def BinarizarImagenes(imagenes):
    for i in range(len(imagenes)):
        imagenes[i] = cv2.cvtColor(imagenes[i], cv2.COLOR_BGR2GRAY)
        _, imagenes[i] = cv2.threshold(imagenes[i], 175, 255, cv2.THRESH_BINARY)

#Detecta los bordes de las imágenes.
def DeteccionBordes(imagenes):
    for i in range(len(imagenes)):
        imagenes[i] = cv2.Canny(imagenes[i],90,150)

#Muestra la imagen de prueba original y sus modificaciones.
def MostrarImagenes(imagenes):
    cv2.imshow("Original", imagenes[0])
    cv2.imshow("Reescalada", imagenes[1])
    cv2.imshow("Binarizada", imagenes[2])
    cv2.imshow("Bordes", imagenes[3])
    cv2.waitKey(0)# Esperar a que se presione una tecla.
    cv2.destroyAllWindows()#Eliminar ventanas.

def GuardarImagenes(imagenes, ruta):
    for raiz, carpetas, archivos in os.walk(ruta):
        i=0
        for archivo in archivos:
            cv2.imwrite(ruta[:-1]+"-borde/"+archivo,imagenes[i])
            i+=1
```

```python
#Importación de las imágenes.
ruta = 'Recursos/music_instruments/'

#Arreglos para guardar las imágenes.
acordeones = []
banjos = []
tambores = []
flautas = []
guitarras = []
armonicas = []
saxofones = []
sitaras = []
tablas = []
violines = []
imagenesPrueba = []
imagenPrueba = 2

GuardarImagenesArreglo(acordeones, ruta+'accordion/')
GuardarImagenesArreglo(banjos, ruta+'banjo/')
GuardarImagenesArreglo(tambores, ruta+'drum/')
GuardarImagenesArreglo(flautas, ruta+'flute/')
GuardarImagenesArreglo(guitarras, ruta+'guitar/')
GuardarImagenesArreglo(armonicas, ruta+'harmonica/')
GuardarImagenesArreglo(saxofones, ruta+'saxophone/')
GuardarImagenesArreglo(sitaras, ruta+'sitar/')
GuardarImagenesArreglo(tablas, ruta+'tabla/')

GuardarImagenesArreglo(violines, ruta+'violin/')
imagenesPrueba.append(violines[imagenPrueba])
```

```python
#Redimensionamiento de las imágenes.
ReescalarImagenes(acordeones)
ReescalarImagenes(banjos)
ReescalarImagenes(tambores)
ReescalarImagenes(flautas)
ReescalarImagenes(guitarras)
ReescalarImagenes(armonicas)
ReescalarImagenes(saxofones)
ReescalarImagenes(sitaras)
ReescalarImagenes(tablas)

ReescalarImagenes(violines)
imagenesPrueba.append(violines[imagenPrueba])
```

```python
#Binarización de las imágenes.
BinarizarImagenes(acordeones)
BinarizarImagenes(banjos)
BinarizarImagenes(tambores)
BinarizarImagenes(flautas)
BinarizarImagenes(guitarras)
BinarizarImagenes(armonicas)
BinarizarImagenes(saxofones)
BinarizarImagenes(sitaras)
BinarizarImagenes(tablas)

BinarizarImagenes(violines)
imagenesPrueba.append(violines[imagenPrueba])
```

```python
#Detectar bordes.
DeteccionBordes(acordeones)
DeteccionBordes(banjos)
DeteccionBordes(tambores)
DeteccionBordes(flautas)
DeteccionBordes(guitarras)
DeteccionBordes(armonicas)
DeteccionBordes(saxofones)
DeteccionBordes(sitaras)
DeteccionBordes(tablas)

DeteccionBordes(violines)
imagenesPrueba.append(violines[imagenPrueba])
```

```python
#Mostrar imágenes y sus transformaciones.
MostrarImagenes(imagenesPrueba)
```

```python
#Guardar imágenes.
GuardarImagenes(acordeones, ruta+'accordion/')
GuardarImagenes(banjos, ruta+'banjo/')
GuardarImagenes(tambores, ruta+'drum/')
GuardarImagenes(flautas, ruta+'flute/')
GuardarImagenes(guitarras, ruta+'guitar/')
GuardarImagenes(armonicas, ruta+'harmonica/')
GuardarImagenes(saxofones, ruta+'saxophone/')
GuardarImagenes(sitaras, ruta+'sitar/')
GuardarImagenes(tablas, ruta+'tabla/')
GuardarImagenes(violines, ruta+'violin/')
```

<h3><strong>Descripción de las técnicas utilizadas</strong></h3>
El fin último es obtener las siluetas de los instrumentos por lo que se aplicaron las siguientes transformaciones:

<ol>
    <li>Redimensionamiento</li>
    Primeramente se hizo una "estandarización" de las imágenes, haciendo que todas tuviesen la misma altura y anchura (512x512[px]) utilizando el método resize de la biblioteca OpenCV.
    <li>Binarización</li>
    Una vez que las imágenes tienen el mismo tamaño se realiza la binarización empezando por hacer un cambio del espacio de colores con el método cvtColor, para que las imágenes se muestren en escala de grises, esto con la finalidad de poder aplicar correctamente el método threshold y que la imagen se represente utilizando solamente con los colores blanco (#FFFFFF) y negro (#000000).
    <li>Detección de bordes</li>
    Finalmente se utiliza el método Canny el cual implementa un algoritmo del mismo nombre para poder detectar los bordes de la imagen, de esta manera nos quedamos solamente con las siluetas de los instrumentos.
</ol>
